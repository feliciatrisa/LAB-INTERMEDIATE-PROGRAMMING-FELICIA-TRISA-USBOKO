```python
import random

secret_words = ["prasmul", "kuliah", "belajar", "ujian", "tugas"]
used_words = []
global score
score = {'win': 0, 'lose': 0}

def main():
    global score
    while True:
        if len(used_words) == len(secret_words):
            print("Wow, you've already guessed all the words correctly!")
            break


        secret = random.choice([word for word in secret_words if word not in used_words])
        used_words.append(secret)

        guess = "_ " * len(secret)
        correct = []
        wrong_guesses = 0
        guessed_letters = []

        while True:
            print("Guess this: ", guess)
            letter = input("Guess a letter: ").lower()

            if letter in guessed_letters:
                print("You've already tried that letter. How about picking another one?")
                continue
            else:
                guessed_letters.append(letter)

            if letter in secret:
                correct.append(letter)
                newguess = ""
                for i in range(0, len(secret)*2, 2):
                    if secret[i//2] == letter:
                        newguess += letter + " "
                    else:
                        newguess += guess[i:i+2]
                guess = newguess
                print("You've nailed it! That letter is correct!")
                if "_" not in guess:
                    print("Well done! You've successfully guessed the correct word!")
                    score['win'] += 1
                    break
            elif not letter.isalpha():
                print("Input a SINGLE LETTER, please!")
            else:
                print("Whoops! That wasn't the right letter.")
                wrong_guesses += 1
                print(f"Wrong guesses left: {6 - wrong_guesses}")

            if wrong_guesses == 6:
                print("Sorry, you're out of guesses. The word was: ", secret)
                score['lose'] += 1
                break

        #Untuk tanya ke user apakah mau melanjutkan game atau tidak
        play_again = input("Would you like to play again? (yes/no) ").lower()
        if play_again != "yes":
            break


    print(f"Final score -> Wins: {score['win']}, Losses: {score['lose']}")

if __name__ == "__main__":
    main()
```
